```python
# Importing all the necessary libraries for Data analysis
# importing numpy,pandas,seabron,matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#cpoying the data and reading it.
df=pd.read_csv('//content/epi_r.csv',header=0)
```

```python
# checking the head/ first 5 values of the data
df.head()
```

```python
#Understanidng the data
df.describe()
```

```python
# knowing the shape of the data
df.shape
```

```python
#Step 1: Removing Duplicate Rows
df.drop_duplicates(inplace=True)
```

```python
#After removing the duplicates the data (checking it)
#shape of data is changed
df.shape
```

```python
#Step 2: Handling Missing Values
df.isnull().sum()


```

```python
#imputation of missing values
for col in ['calories', 'fat', 'protein', 'sodium']:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
#checking for missing value again
df.isnull().sum()
```

Outlier Detection- As seeing the data there seems to be no outliers present as the calories count can be a very subjective value and can vary from oil amount, quantity of vegetables, cheese and other ingrediant taken to prepare the item.
High-calorie recipes often contain ingredients like oils, butter, or cheese, which naturally contribute to higher calorie content.
Low-calorie recipes might include lighter ingredients like vegetables or lean proteins.

**Exploratory Data Analysis**

```python
# Finding the correlation
#We'll calculate the correlations between the rating and nutritional columns (calories, fat, protein, sodium).
#Correlation between rating and nutritional values
corr=df[['rating', 'calories', 'fat', 'protein', 'sodium']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Box plot for calories
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['calories'], color='green')
plt.title('Box Plot of Calories in Recipes')
plt.xlabel('Calories')
plt.show()
```

****

```python
# Pair plot of key variables
plt.figure(figsize=(15, 10))
sns.pairplot(df[['rating', 'calories', 'protein', 'fat', 'sodium']], diag_kind='kde')
plt.suptitle('Pair Plot of Nutritional Information and Rating', y=1.02)
plt.show()
```

```python
#Visualize the Distribution of Recipe Ratings
#First, we'll explore the distribution of recipe ratings using a histogram.

import matplotlib.pyplot as plt

# Plotting the distribution of ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=20, color='blue')
plt.title('Distribution of Recipe Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot between calories and rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='calories', y='rating', data=df)
plt.title('Calories vs Rating')
plt.xlabel('Calories')
plt.ylabel('Rating')
plt.xlim(100,1200)
plt.ylim(1,5)

plt.show()
```

```python
# Analyze Top Ingredients in Highly Rated Recipes
# Select highly-rated recipes (rating >= 4)
highly_rated = df['rating'] >= 4
highly_rated
```

```python
highly_rated.value_counts()
```

```python
df.columns
```

```python
# Analyze top ingredients in highly rated recipes
ingredient_cols = df.columns[15:]

# Summing ingredient usage in highly rated recipes
top_ingredients =df[highly_rated][ingredient_cols].sum().sort_values(ascending=False).head(35)

# Bar plot of top ingredients in highly rated recipes
plt.figure(figsize=(12, 8))
sns.barplot(x=top_ingredients.values, y=top_ingredients.index, palette='viridis')
plt.title('Top Ingredients in Highly Rated Recipes')
plt.xlabel('Count of Ingredient Occurrence')
plt.ylabel('Ingredients')
plt.show()
```

Creativity to find the proportion of non veg and veg choices.

```python
import matplotlib.pyplot as plt

veg_count = len(df[df['vegetarian'] == 1.0])
nonveg_count = len(df[df['vegetarian'] == 0.0])


labels = ['Vegetarian', 'Non-Vegetarian']
sizes = [veg_count, nonveg_count]
colors = ['green', 'red']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, startangle=90)
plt.axis('equal')#equal ratio
plt.title('Proportion of Vegetarian and Non-Vegetarian Dishes')
plt.show()
```
